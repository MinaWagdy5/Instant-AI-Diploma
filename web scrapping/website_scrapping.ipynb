```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define URL 
url = 'https://www.yallakora.com/Match-Center/?date=18/21/2023'
```

```python
response = requests.get(url)
html_content = response.content
```

```python
print(html_content)
```

```python
# create beautifulsoup object to parse html content 
soup = BeautifulSoup(html_content , 'html.parser')
```

```python
soup
```

```python
# find all match elements in the page
championships = soup.findAll('div' , class_='matchCard')
```

```python
match_details = []
for i in range(len(matches2)):
    teamA = matches2[i].find('div' , {'class':'teamA'}).text.strip()
    teamB = matches2[i].find('div' , {'class':'teamB'}).text.strip()

    results = matches2[i].find('div' , {'class':'MResult'}).find_all('span' , {'class':'score'})

    score = f'{results[0].text.strip()} . {results[1].text.strip()}'
    time = matches2[i].find('div' , {'class':'MResult'}).find('span' , {'class':'time'}).text.strip()

    print(teamA , score, teamB  , time)
```

```python
for i in range(len(championships)):
    print(championships[i].contents[1].find('h2').text.strip())
    matches = championships[i].contents[3].find_all(class_='liItem')

    for j in range(len(matches)):
        teamA = matches[j].find('div' , class_='teamA').text.strip()
        teamB = matches[j].find('div' , class_='teamB').text.strip()

        results = matches[j].find('div' , class_='MResult').find_all('span' , class_='score')

        score = f'{results[0].text.strip()} . {results[1].text.strip()}'
        time = matches[j].find('div' , class_='MResult').find('span' , class_='time').text.strip()
        print(teamA , score, teamB  , time)

        #store data in a file
        matches_file = open('myfile.txt' , 'a')   # a:append , w:write , r:read
        matches_file.write(f'{teamA} {score} {teamB} {time}\n')
        matches_file.write('=======================\n')

    print('='*20)
```
